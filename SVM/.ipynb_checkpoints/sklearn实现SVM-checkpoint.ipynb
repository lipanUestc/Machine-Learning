# 1、Load Data

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()

print(dir(data))  
#查看data所有的属性和方法

#print(data.DESCR) 
#查看数据集简介

print(data.data.shape)
#数据集形状，150样本，4属性

data.feature_names
```

# 2、Scatter Plot 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.squeeze(data.data[:,0:1])
x2 = np.squeeze(data.data[:,1:2]) #这里要squeeze一下，要传入数组而不是矩阵
plt.scatter(x1,x2,s=75,c=data.target,alpha=0.5)
plt.show()
```

# 3、Model

```python
from sklearn import svm
import pandas as pd
```

```python
#model = svm.SVC(C=0.1, kernel='linear', decision_function_shape='ovr')
model = svm.SVC(C=0.8, kernel='rbf', gamma=20, decision_function_shape='ovr')
#SVC就是SVM的classifier分类模型，   SVR就是regression回归模型，用于预测具体值
model.fit(data.data[:,:2],data.target)


result=[]
Error =0
for i in range(150):
    ret = int(model.predict(data.data[i:i+1,:2]))
    jug = abs(ret-data.target[i])
    if(jug!=0):
        Error+=1
        result.append(0)
    else:
        result.append(1)
print("Error Rate:",Error/150)
```

```python
plt.scatter(x1,x2,s=75,c=result,alpha=0.5)
plt.show()
```
